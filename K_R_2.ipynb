```python
#task 2
from surprise import NormalPredictor
from surprise import Dataset
from surprise import Reader
from surprise import BaselineOnly
from surprise import accuracy
from surprise.model_selection import train_test_split

reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(rating[['user_id', 'book_id', 'rating']], reader)
trainset, testset = train_test_split(data, test_size=.25)

Base = BaselineOnly()
Base.fit(trainset)
predictions = Base.test(testset)
accuracy.rmse(predictions)
```

```python
Normal = NormalPredictor()
Normal.fit(trainset)
predictions = Normal.test(testset)
accuracy.rmse(predictions)
```

```python
from surprise.model_selection import KFold
from surprise.model_selection import LeaveOneOut
from surprise.model_selection import RepeatedKFold

kf = KFold(n_splits=5)
loo = LeaveOneOut(n_splits=5)
rkf = RepeatedKFold(n_splits=3)

kf_results = []
loo_results = []
rkf_results = []
algo = BaselineOnly()

for trainset, testset in kf.split(data):
    algo.fit(trainset)
    predictions = algo.test(testset)
    kf_results.append(accuracy.rmse(predictions, verbose=True))

for trainset, testset in loo.split(data):  
    algo.fit(trainset)
    predictions = algo.test(testset)
    loo_results.append(accuracy.rmse(predictions, verbose=True))
    
for trainset, testset in rkf.split(data):  
    algo.fit(trainset)
    predictions = algo.test(testset)    
    rkf_results.append(accuracy.rmse(predictions, verbose=True))
    
print('\n KFold results:', kf_results)
print("\n LeaveOneOut results:",  loo_results) 
print("\n 'RepeatedKFold results:", rkf_results)
```
